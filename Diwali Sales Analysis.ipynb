```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\ADMIN\Desktop\Python project\Diwali Sales Data.csv',encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename (columns = {'Marital_Status' : 'Marriage Tag'})
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.info ()
```

```python
df.dropna (inplace=True)
```

```python
df.shape
```

```python
pd.isnull (df).sum()
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.describe()
```

```python
df[['Age', 'Orders', 'Amount']].describe ()
```

```python
ax = sns.countplot ( x = 'Gender' , data = df )

for bars in ax.containers :
    ax.bar_label (bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot (x = 'Gender', y = 'Amount', data=sales_gen)
```

```python
From above graphs we can see that female buyers are more & their purchasing power is greater than males.
```

```python
sns.countplot (data = df,x = 'Age Group', hue = 'Gender')
```

```python
sns.countplot (data = df,x = 'Age Group')
```

```python
ax = sns.countplot (data = df,x = 'Age Group', hue = 'Gender')

for bars in ax.containers :
    ax.bar_label (bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum.sort_values(by='Amount',ascending=False)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot (x = 'Age Group', y = 'Amount', data=sales_age)

```

```python
Above graph shows that female with age agroup of 26-35 yrs purchase more than other.
```

```python
sales_state =df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'State', y = 'Orders', data=sales_state)
```

```python
sales_state =df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'State', y = 'Amount', data=sales_state)
```

```python
Above graph shows that both highest number of orders & highest amount of orders placed from staes like UP,Maharashtra followed by Karnataka.
```

```python
ax = sns.countplot ( x = 'Marital_Status' , data = df )

for bars in ax.containers :
    ax.bar_label (bars)
```

```python
ax = sns.countplot ( x = 'Marital_Status' , data = df )

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers :
    ax.bar_label (bars)
```

```python
sales_state =df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x = 'Marital_Status', y = 'Amount', data=sales_state, hue = 'Gender')
```

```python
Above graph shows that both highest number of orders palced by married people & that too by female customers.
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot ( x = 'Occupation' , data = df )
for bars in ax.containers :
    ax.bar_label (bars)
```

```python
sales_state =df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Occupation', y = 'Amount', data=sales_state)
```

```python
Above graph shows that both highest number of orders & highest amount of orders placed by the people from Occupation like IT Sector,Healthcare & Aviation.
```

```python
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot ( x = 'Product_Category' , data = df )
for bars in ax.containers :
    ax.bar_label (bars)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
sales_state =df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head (10)
sns.barplot(x = 'Product_Category', y = 'Amount', data=sales_state)
```

```python
Above graph shows that both highest number of orders placed for categories like Clothing,Food followed by electronics. In terms of amount highest orders placed for 
categories like Food,Clothing followed by electronics.
```

```python
sns.set(rc={'figure.figsize':(20,5)})
sales_state =df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head (10)
sns.barplot(x = 'Product_ID', y = 'Orders', data=sales_state)
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot (kind='bar')
```

```python
In Conclusion maximum orders palced by Married people, Female customers having occupation like IT,Healthcare,Aviation for product categories
like Food,Clothing & Elecetronics.
```
